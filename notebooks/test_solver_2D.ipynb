```python
import sys
import os

sys.path.append(os.path.abspath(".."))

import matplotlib.pyplot as plt
```

```python
from src.solvers import AB2AM2Solver2D
from src.models import SchnakenbergModel

params = {
    "a": 0.1,
    "b": 0.9,
    "gamma": 2.0,
    "d": 20.0
}

solver = AB2AM2Solver2D(
    model=SchnakenbergModel,
    params=params,
    Lx=100,
    Ly=100,
    Nx=100,
    Ny=200,
    T=50,        # start smaller for testing
    dt=0.1,
    save_every=10
)
```

```python
t, U, V, x, y, X, Y = solver.run(p_type="random")

print("Ran successfully:", U.shape)
```

```python
plt.figure(figsize=(12, 5))

# U
plt.subplot(1, 2, 1)
plt.contourf(X, Y, U[-1], cmap='viridis', levels=50)
plt.colorbar(label='U')
plt.title("Final U")

# V
plt.subplot(1, 2, 2)
plt.contourf(X, Y, V[-1], cmap='plasma', levels=50)
plt.colorbar(label='V')
plt.title("Final V")

plt.tight_layout()
plt.show()
```
